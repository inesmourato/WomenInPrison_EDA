```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import warnings
warnings.filterwarnings('ignore')
```

```python
merged_edu=pd.read_csv('/Users/inesmourato/Documents/AllWomen/FinalProject/merged_edu_clean.csv')
```

```python
merged_edu
```

```python
plt.figure(figsize=(12, 6))
filtered_df=merged_edu[merged_edu['gender']!='both']
sns.lineplot(data=filtered_df, x='year', y='rate_basic', hue='gender')
plt.title('Evolution of Enrollment Rates in Primary Education Over Time with Gender Differences')
plt.xlabel('Year')
plt.ylabel('Enrollment Rates')
plt.ylim(100, None) 
plt.legend(title='Gender')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
filtered_df=merged_edu[merged_edu['gender']!='both']
sns.lineplot(data=filtered_df, x='year', y='rate_high_school', hue='gender')
plt.title('Evolution of Enrollment Rates in Secondary Education Over Time with Gender Differences')
plt.xlabel('Year')
plt.ylabel('Enrollment Rates')
plt.ylim(0, None) 
plt.legend(title='Gender')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
filtered_df=merged_edu[merged_edu['gender']!='both']

sns.lineplot(data=filtered_df, x='year', y='rate_superior', hue='gender')
plt.title('Evolution of Enrollment Rates in Superior Education Over Time with Gender Differences')
plt.xlabel('Year')
plt.ylabel('Enrollment Rates')
plt.ylim(0, None) 
plt.legend(title='Gender')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
filtered_df=merged_edu[merged_edu['gender']!='both']
sns.lineplot(data=filtered_df, x='year', y='rate_dropout', hue='gender')
plt.title('Evolution Early Dropout Rates in Basic Education Over Time with Gender Differences')
plt.xlabel('Year')
plt.ylabel('Dropout Rates')
plt.legend(title='Gender')
plt.ylim(0, None) 

plt.show()
```

```python
sns.pairplot(merged_edu[["rate_basic", "rate_high_school", "rate_superior", "rate_dropout"]], diag_kind="kde")

plt.show()
```

```python
filtered_df = merged_edu[merged_edu['gender'] != 'both']

# Select only the numeric columns for correlation
corr_matrix = filtered_df[['rate_basic', 'rate_high_school', 'rate_superior', 'rate_dropout']].corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title('Correlation Matrix')
plt.show()
```

```python
merged_empl=pd.read_csv('/Users/inesmourato/Documents/AllWomen/FinalProject/merged_employment.csv')
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=merged_empl[merged_empl['gender'] != 'HM'], x='year', y='salary_gross', hue='gender', palette={'H': 'blue', 'M': 'red'}, ci=None)

plt.title('Average Gross Salary by Gender Over Time')
plt.xlabel('Year')
plt.ylabel('Gross Salary ()')
plt.xticks(rotation=45)
plt.legend(title='Gender', labels=['Men', 'Women'])

plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=merged_empl[merged_empl['gender'] == 'HM'], x='salary_gross', y='percentage', color='purple')

plt.title('Wage Gap vs. Gross Salary Levels')
plt.xlabel('Gross Salary ()')
plt.ylabel('Wage Gap Percentage')
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=merged_empl[merged_empl['gender'] != 'HM'], x='gender', y='salary_gross', palette={'H': 'blue', 'M': 'pink'})

plt.title('Salary Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Gross Salary ()')
plt.xticks(ticks=[0, 1], labels=['Men', 'Women'])

plt.show()
```
